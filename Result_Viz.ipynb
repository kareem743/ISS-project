```python
import numpy as np
import matplotlib.pyplot as plt


file_sizes_mb = [10, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1024]
transfer_times = [0.1432, 0.7945, 1.4709, 2.2452, 3.3234, 5.4014, 6.5026, 6.1713, 8.2723, 9.9120, 10.0973]


coefficients = np.polyfit(file_sizes_mb, transfer_times, 1)
line_fit = np.poly1d(coefficients)


predicted_times = line_fit(file_sizes_mb)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(file_sizes_mb, transfer_times, color='blue', marker='o')
plt.plot(file_sizes_mb, predicted_times, color='red', linestyle='--')


plt.title("File Size vs Transfer Time")
plt.xlabel("File Size (MB)")
plt.ylabel("Transfer Time (seconds)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()


plt.tight_layout()
plt.show()
```
